```python
import networkx as nx
import matplotlib.pyplot as mpl
import plotly.plotly as plt
import plotly.graph_objs as go
import math
import json
import pprint

%matplotlib notebook
```

```python
retweeter_to_owner_relation = {}
```

```python
with open("data.json") as tweets:
    for tweet in tweets:
        tweet_json = json.loads(tweet)
        if("retweeted_status" in tweet_json):
            if(tweet_json["user"]["screen_name"] in retweeter_to_owner_relation):
                retweeter_to_owner_relation[tweet_json["user"]["screen_name"]] += tweet_json["retweeted_status"]["user"]["screen_name"]
            else:
                retweeter_to_owner_relation[tweet_json["user"]["screen_name"]] = [tweet_json["retweeted_status"]["user"]["screen_name"]]
```

```python
print(retweeter_to_owner_relation)
G = nx.DiGraph()
for retweeter in retweeter_to_owner_relation.keys():
    G.add_edge(retweeter,retweeter_to_owner_relation[retweeter])
```

```python
mpl.figure()
nx.draw_networkx(G)
```
